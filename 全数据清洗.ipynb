```python
import pandas as pd
import numpy as np
```

```python
#颜色映射为RGB
color_rgb={1:[255,255,255],2:[0,0,255],3:[160,82,45],4:[160,32,240],5:[255,0,0],6:[255,255,0],7:[192,192,192],8:[0,255,0],9:[233,233,216],0:[0,0,0]}
```

```python
df=pd.read_csv('dataset_1.csv')
```

```python
color_rgb[0]
```

```python
item_id=df['衣服ID']
type(item_id)
```

```python
df['R']=0
df['G']=0
df['B']=0
df['花纹']=0
df['颜色深度']=0
for i in range(len(df)):#转换色彩RGB与颜色深度、花纹
    id=item_id[i]
    id=list(id[-4:-1])
    df['花纹']=id[0]
    df['颜色深度']=id[2]    
    rgb=color_rgb[int(id[1])]
    df['R']=rgb[0]
    df['G']=rgb[1]
    df['B']=rgb[2]
```

```python
order_data=pd.read_excel('唯衣POC.xlsx',sheet_name=0)
item_id=pd.read_excel('唯衣POC.xlsx',sheet_name=3)
```

```python
item_id=item_id.drop(0)
order_data=order_data.drop(0)
```

```python
item_id["条码"]=[i[:9] for i in item_id["条码"]]
```

```python
order_data=order_data[order_data['订单总金额']>0]
```

```python
user_item=pd.merge(order_data,item_id,left_on=['订单所含商品条码ID'],right_on=['条码id'],how='left')
```

```python
user_item["buys"]=1
user_item['下订单VIP']=user_item['下订单VIP'].apply(np.round)
user_item.groupby(["下订单VIP","条码"])["buys"].sum().to_csv('alluib_without_color.csv')
user_item[['下订单VIP','条码','buys']]
```
