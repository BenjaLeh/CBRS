```python
import matplotlib.pyplot as plt


def dv_encoder(data):#oneHot编码
    # Importing the dataset
    X=data.iloc[:,:].values#转化为numpy
    
    from sklearn.preprocessing import LabelEncoder, OneHotEncoder
    for i in range(1,15):
        labelencoder_X = LabelEncoder(encoding='utf-8')
        X[:, i] = labelencoder_X.fit_transform(X[:,i])
        
    onehotencoder = OneHotEncoder(categorical_features = list(range(1,10)))
    y = onehotencoder.fit_transform(X).toarray()
    
    return y
```

```python
Bitems=pd.get_dummies(items,columns=cols)
print(Bitems)
Bitems=Bitems.drop(['衣服ID'],axis=1)
```

```python
from sklearn.neighbors import NearestNeighbors
""" 
NearestNeighbors用到的参数解释 
n_neighbors=5,默认值为5，表示查询k个最近邻的数目 
algorithm='auto',指定用于计算最近邻的算法，auto表示试图采用最适合的算法计算最近邻 
fit(X)表示用X来训练算法 
"""  
nbrs = NearestNeighbors(n_neighbors=322, algorithm="ball_tree").fit(Bitems)  
#返回距离每个点k个最近的点和距离指数，indices可以理解为表示点的下标，distances为距离  
distances, indices = nbrs.kneighbors(Bitems)  
print (indices ) 
print(distances)  
```
