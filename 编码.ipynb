```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('dataset_1.csv')
print(df.columns)
df=df.drop(['色相','Unnamed: 0'],axis=1)
df
```

```python
#产生userID集与itemID集
items=df.drop(['客户ID'],axis=1).drop_duplicates(['衣服ID'])#物品特征表
itemsID=items['衣服ID']#物品id表
userID=df['客户ID'].drop_duplicates()#去重
b=items.pop('衣服ID')
items.insert(0,'衣服ID',b)
itemsID
```

```python
#需要onehot编码的列名
cols=[  '衣服单价',  '款式流水号', '颜色深度', '年份', '季节',
       '小类号码', '颜色式样', '颜色大类', '颜色名称', '品名', '大类', '中类', '品牌定义廓形', '领型', '袖长',
       '系列', '尺寸组', '标准价', '色温', '色明度', '色纯度', '对比度', '色调', '廓形', '宽松度',
       '长度', '款设图-大小复杂度', '款设图-规则性', '款设图-曲直', '材质-垂感', '整体曲直', '量感', '场合',
       '领型-形状', '领型-曲直', '领型-量感', '裙型裤型']
```

```python
Bitems=pd.get_dummies(items,columns=cols)
print(Bitems)
Bitems.to_csv('item_feature.csv')
Bitems=Bitems.drop(['衣服ID'],axis=1)
```

```python
def compute_consin(item):
    x=np.zeros((322,322))
    for i in range(322):
        A=item[i]
        A=A[:,None]
        for j in range(322):
            B=item[j]
            B=B[:,None].T
            num = np.dot(B,A) #若为行向量则 A * B.T           
            denom = np.linalg.norm(A) * np.linalg.norm(B)  
            cos = num / denom #余弦值 
            sim = 0.5 + 0.5 * cos #归一化
            x[i,j]=sim
    return x
Bitems=Bitems.as_matrix(columns=None)  
xx=compute_consin(Bitems)

```

```python
distances=pd.DataFrame(xx,index=itemsID,columns=itemsID)
print(distances)
```

```python
#编写userID_itemID矩阵
user_item=pd.DataFrame(index=userID,columns=itemsID)
```

```python

from sklearn.model_selection import train_test_split

X_train, X_test,= train_test_split(df, test_size = 0.3, random_state = 0)
```

```python
    user_item=user_item.fillna(0)
    a=X_train.groupby('客户ID')
    for user,grouped in a:
        for item in grouped['衣服ID']:
            user_item.loc[user, item]=1       
```

```python
recommender={}#推荐列表    
a=X_train.groupby('客户ID')
for user,grouped in a:
    x={}#存放推荐物品字典       
    for item in itemsID:
        if item in grouped['衣服ID']:
            continue
        elif len(grouped['衣服ID'])==0:
            x[item]=0
        else:    
            x[item]=max(distances.loc[item,grouped['衣服ID']])
    recommender[user]=x
recommender
```

```python
#排序
for d in recommender:
    a=recommender[d]
    recommender[d]=sorted(a.items(),key=lambda a:a[1],reverse=True)    
```
