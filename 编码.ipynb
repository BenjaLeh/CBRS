```python
def compute_consin(item):
    x=np.zeros((322,322))
    for i in range(322):
        A=item[i]
        A=A[:,None]
        for j in range(322):
            B=item[j]
            B=B[:,None].T
            num = np.dot(B,A) #若为行向量则 A * B.T           
            denom = np.linalg.norm(A) * np.linalg.norm(B)  
            cos = num / denom #余弦值 
            sim = 0.5 + 0.5 * cos #归一化
            x[i,j]=sim
    return x
#Bitems=Bitems.as_matrix(columns=None)  
xx=compute_consin(Bitems)                   
```
